```python
import numpy as np
import pandas as pd
%pylab inline
```

```python
log = pd.read_csv('log.csv')
l = log[:10000]
```

# Position

```python

figure(figsize=(10,5))

subplot(1,2,1)
plot(l.x, -l.z)
xlabel('x')
ylabel('z')
title('Z Position')

subplot(1,2,2)
plot(l.x,l.y)
xlabel('x')
ylabel('y')
title('Y position')

tight_layout()
```

# Speed

```python
figure(figsize=(15,5))

subplot(1,3,1)
plot(l.index, l.vx)
xlabel('step')
ylabel('vx')
title('Horizontal x speed')

subplot(1,3,2)
plot(l.index, l.vy)
xlabel('step')
ylabel('vy')
title('Horizontal y speed')

subplot(1,3,3)
plot(l.index, -l.vz)
xlabel('step')
ylabel('vz')
title('Vertical speed')

tight_layout()
```

# Angle of attack

```python
figure(figsize=(15,5))

subplot(1,4,1)
plot(l.index, np.rad2deg(l.aoa1))
xlabel('step')
ylabel('aoa')
title('Angle of attack of left flap')

subplot(1,4,2)
plot(l.index, np.rad2deg(l.aoa2))
xlabel('step')
ylabel('aoa')
title('Angle of attack of right flap')

subplot(1,4,3)
plot(l.index, np.rad2deg(l.aoa3))
xlabel('step')
ylabel('aoa')
title('Angle of attack of left drift')

subplot(1,4,4)
plot(l.index, np.rad2deg(l.aoa4))
xlabel('step')
ylabel('aoa')
title('Angle of attack of right drift')

tight_layout()
```

# Orientation

```python
figure(figsize=(15,5))

subplot(1,3,1)
plot(np.rad2deg(l.r))
title('Roll')

subplot(1,3,2)
plot(np.rad2deg(l.p))
title('Pitch')

subplot(1,3,3)
plot(np.rad2deg(l.y))
title('Yaw')

tight_layout()

```
