```python
import numpy as np
import pandas as pd
%pylab inline
```

```python
log = pd.read_csv('log.csv')
```

# Position

```python
l = log[:1000]

plot(l.x, -l.z)
xlabel('x')
ylabel('z')
title('Position')
```

# Speed

```python


figure(figsize=(10,5))

subplot(1,2,1)
plot(l.index, l.vx)
xlabel('step')
ylabel('vx')
title('Horizontal speed')

subplot(1,2,2)
plot(l.index, -l.vz)
xlabel('step')
ylabel('vz')
title('Vertical speed')

tight_layout()
```

# Pitch

```python
plot(np.rad2deg(l.p))
title('Pitch')
```
