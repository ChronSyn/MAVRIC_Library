```python
import numpy as np
import pandas as pd
%pylab inline
```

```python
log = pd.read_csv('log.csv')
```

# Position

```python
l = log[:1000]

plot(l.x, -l.z)
xlabel('x')
ylabel('z')
title('Position')
```

# Speed

```python


figure(figsize=(10,5))

subplot(1,3,1)
plot(l.index, l.vx)
xlabel('step')
ylabel('vx')
title('Horizontal speed')

subplot(1,3,2)
plot(l.index, -l.vz)
xlabel('step')
ylabel('vz')
title('Vertical speed')

subplot(1,3,3)
plot(l.index, np.rad2deg(np.arctan2(l.vz,l.vx)+l.p))
xlabel('step')
ylabel('aoa')
title('Angle of attack')

tight_layout()
```

# Pitch

```python
plot(np.rad2deg(l.p))
title('Pitch')
```

```python
plot(l.index,l.pt)
```

```python
plot(l.index,l.pt/(l.vx*l.vx+l.vz*l.vz))
```
