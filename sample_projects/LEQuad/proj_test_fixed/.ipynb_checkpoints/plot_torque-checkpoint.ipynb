```python
import numpy as np
import pandas as pd
%pylab inline
```

```python
log = pd.read_csv('log.csv')
l = log[:1000000]
```

# Position

```python

figure(figsize=(10,5))

subplot(1,2,1)
plot(l.angle, l.torque)
xlabel('x')
ylabel('z')
title('Z Position')

subplot(1,2,2)
plot(l.x,l.y)
xlabel('x')
ylabel('y')
title('Y position')

tight_layout()
```

# Speed

```python
figure(figsize=(15,5))

subplot(1,3,1)
plot(l.t, l.vx)
xlabel('time (s)')
ylabel('vx')
title('Horizontal x speed')

subplot(1,3,2)
plot(l.t, l.vy)
xlabel('time (s)')
ylabel('vy')
title('Horizontal y speed')

subplot(1,3,3)
plot(l.t, -l.vz)
xlabel('time (s)')
ylabel('vz')
title('Vertical speed')

tight_layout()
```

# Orientation

```python
figure(figsize=(15,5))

subplot(1,3,1)
plot(l.t,np.rad2deg(l.roll))
xlabel('Time (s)')
ylabel('Roll angle (In degrees)')
title('Roll')

subplot(1,3,2)
plot(l.t,np.rad2deg(l.pitch))
xlabel('Time (s)')
ylabel('Pitch angle (In degrees)')
title('Pitch')

subplot(1,3,3)
plot(l.t,np.rad2deg(l.yaw))
xlabel('Time (s)')
ylabel('Yaw angle (In degrees)')
title('Yaw')

tight_layout()

```
